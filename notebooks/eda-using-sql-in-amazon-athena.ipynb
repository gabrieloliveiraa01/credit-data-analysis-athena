## 1. Introduction

This project presents an exploratory data analysis of a dataset containing information about credit card customers. The dataset includes 2,564 records and 13 variables describing customer demographics, financial characteristics, and credit card usage behavior.

The variables analyzed include age, gender, number of dependents, education level, marital status, annual salary range, card type, number of financial products, months of inactivity, credit limit, total transaction value over the last 12 months, and total number of transactions during the same period.

The main objective of this analysis is to explore patterns in customer behavior and understand how demographic and financial characteristics may relate to credit usage and spending behavior.

## 2. Data Structure and Analysis Scope

The dataset contains information about 2,564 customers and includes variables related to their demographic profile, financial characteristics, and credit card usage behavior. Each row represents a single customer, while each column represents a specific attribute describing that customer.

These variables allow us to analyze patterns in customer behavior, spending habits, and credit usage.

The dataset includes the following columns:

>**age** = customer's age, used to understand the age distribution of the client base.
>
>**gender** = customer's gender (F or M), which helps analyze differences in behavior between male and female customers.
>
>**dependents** = number of dependents associated with the customer, which may influence financial responsibilities and spending patterns.
>
>**education_level** = customer's level of education, allowing analysis of whether education level is associated with spending behavior or credit usage.
>
>**marital_status** = customer's marital status (such as single, married, or divorced), which can provide insights into demographic patterns.
>
>**annual_salary** = customer's annual salary range, used to understand the financial profile of the customers and analyze how income levels relate to spending and credit limits.
>
>**card_type** = type of credit card used by the customer (for example Blue, Silver, Gold, or Platinum), which may be associated with different spending patterns and credit limits.
>
>**products_qty** = number of financial products the customer has with the bank, indicating the level of engagement with the institution.
>
>**interactions_12m** = number of interactions or contacts with the bank during the last 12 months, which may indicate customer activity and engagement.
>
>**months_inactive_12m** = number of months in which the customer was inactive during the last year, helping identify customers who may be disengaged.
>
>**credit_limit** = total credit limit available to the customer, representing their borrowing capacity.
>
>**transactions_value_12m** = total monetary value of credit card transactions during the last 12 months, which helps measure spending behavior.
>
>**transactions_qty_12m** = total number of transactions performed during the last 12 months, allowing analysis of customer activity frequency.

Together, these variables provide a comprehensive view of customer demographics, financial status, and credit card usage behavior, enabling deeper analysis through SQL queries to identify patterns and insights related to credit usage and customer behavior.

## 4. Exploratory Data Analysis (EDA)

The first step of the analysis consists of exploring the dataset in order to understand its structure, the available variables, and the overall distribution of the data. This stage is important to verify the number of records, identify the types of variables present in the dataset, and observe basic characteristics of the data before performing more detailed analyses.

Initially, queries are used to visualize a sample of the dataset and confirm that the data was correctly loaded into the table. It is also possible to count the total number of customers and inspect the structure of the table, including the names and data types of each column.

---

### How many customers are in the dataset?

**Query:**

>SELECT COUNT(*) FROM credit_customers;

**Answer:**

>2564 rows

**Observation:**

This result represents the total number of customers available in the dataset and serves as the starting point for the analysis.

---

### What is the gender distribution of customers?

**Query:**

>SELECT gender,
>
>COUNT(*) AS total_customers
>
>FROM credit_customers
>
>GROUP BY gender;

**Answer:**
>Females: 1001
>
>Males: 1563
>
>This query shows how customers are distributed by gender, allowing us to understand the demographic composition of the dataset.

---

### How are customers distributed across different age groups?

**Query:**
>SELECT age,
>
>COUNT(*) AS total_customers
>
>FROM credit_customers
>
>GROUP BY age
>
>ORDER BY age;

**Answer:**
>50+ years: 998
>
>40-49 years: 742
>
>30-39 years: 717
>
>18-29 years: 107
>
>There are no customers under the age of 18 in the dataset.
>
>This analysis helps visualize how customers are distributed across different age groups and identify the most common age ranges within the dataset. The results show that the majority of customers are **over 30 years old**, particularly in the **40–49 and 50+ age groups**. In contrast, the smallest portion of customers is **under 29 years old**, indicating that younger individuals represent a relatively small share of the customer base.


---

### What is the income distribution of customers?

**Query:**
>SELECT annual_salary,
>
>COUNT(*) AS total_customers
>
>FROM credit_customers
>
>GROUP BY annual_salary
>
>ORDER BY total_customers DESC;

**Answer:**
>701 customers earn less than \$40K per year
>
>467 customers earn between \$40K and \$60K
>
>451 customers earn between \$60K and \$80K
>
>488 customers earn between \$80K and \$120K
>
>222 customers earn more than \$120K
>
>235 customers have no salary information (NA)
>
>This query shows how customers are distributed across different salary ranges, helping to understand the financial profile of the client base. Most customers earn less than \$40K per year, while a significant portion of the dataset is concentrated between \$40K and \$120K. A smaller group earns more than \$120K, and some records do not contain salary information.

---

### What is the average credit limit among customers?

**Query:**
>SELECT AVG(credit_limit) AS average_credit_limit
>
>FROM credit_customers;

**Answer:**
>\$9,072.72
>
>This query calculates the average credit limit available among all customers in the dataset.

---

### Which customers have the highest credit limits?

**Query:**
>SELECT credit_limit, gender, annual_salary, card_type
>
>FROM credit_customers
>
>ORDER BY credit_limit DESC
>
>LIMIT 10;

**Answer:**
>The results show that the highest credit limits in the dataset are around 34,500. Most of the customers with the highest limits are male, and many of them have annual salaries between \$80K and \$120K or above \$120K.
>
>Regarding card type, the majority of these customers use Blue, Silver, or Gold cards, with a smaller presence of Platinum cards. Some records also contain "na" in the salary field, indicating that salary information is not available for those customers.

---

### Which card type has the highest average spending?

**Query:**
>SELECT card_type,
>
>AVG(transactions_value_12m) AS average_spending
>
>FROM credit_customers
>
>GROUP BY card_type
>
>ORDER BY average_spending DESC;

**Answer:**
>The results show that customers with Gold cards have the highest average spending, with an average of approximately \$2040.23. They are followed by Silver card users, with an average spending of about \$1883.25, and Blue card users, with an average of \$1816.73.

Customers with Platinum cards present the lowest average spending in this dataset, with an average of approximately 1543.23. This suggests that, within this dataset, Gold card holders tend to have the highest transaction values over the analyzed period.

---

### Who are the customers with the highest annual spending?

**Query:**
>SELECT transactions_value_12m, gender, annual_salary, card_type
>
>FROM credit_customers
>
>ORDER BY transactions_value_12m DESC
>
>LIMIT 10;

**Answer:**
>The results show the customers with the highest transaction values over the last 12 months. The highest spending recorded is approximately \$4776.58, and all customers in the top 10 use the Blue card.
>
>Regarding salary distribution, these customers belong to different income ranges, including less than \$40K, \$40K–\$60K, \$60K–\$80K, \$80K–\$120K, and above \$120K, while some records contain "na", indicating that salary information is not available.
>
>This suggests that high spending is not limited to a single salary range and that customers from different income groups can appear among the highest spenders in the dataset.

---

### Is there a difference in spending between male and female customers?

**Query:**
>SELECT gender,
>
>AVG(transactions_value_12m) AS average_spending
>
>FROM credit_customers
>
>GROUP BY gender;

**Answer:**
>The results show that female customers (F) have an average annual transaction value of approximately \$1839.62, while male customers (M) have an average spending of about \$1807.97.

This indicates that, within this dataset, female customers have a slightly higher average spending compared to male customers, although the difference between the two groups is relatively small.

---

### Are there inactive customers with high credit limits?

**Query:**
>SELECT months_inactive_12m, credit_limit, card_type
>
>FROM credit_customers
>
>WHERE months_inactive_12m >= 6
>
>ORDER BY credit_limit DESC;

**Answer:**
>The results show customers who have been inactive for 6 months but still maintain relatively high credit limits. The highest credit limit among these inactive customers is approximately \$30,948.84.

Most of the customers in this group use the Blue card, with only a small presence of Silver cards. This may indicate that some customers with high available credit are not actively using their cards, which could represent a potential customer disengagement or inactivity pattern in the dataset.

### How does customer education level influence average credit card spending?

**Query:**
>SELECT education_level,
>
>COUNT(*) AS total_customers,
>
>AVG(transactions_value_12m) AS average_spending
>
>FROM credit_customers
>
>GROUP BY education_level
>
>HAVING education_level != 'unknown'
>
>ORDER BY average_spending DESC;

**Answer:**
>Graduate degree: 261 customers — average spending \$1862.24
>
>Doctorate: 235 customers — average spending \$1828.66
>
>Master's degree: 810 customers — average spending \$1824.61
>
>No formal education: 367 customers — average spending \$1819.91
>
>High school: 545 customers — average spending \$1790.15
>
>The results show that customers with higher education levels tend to present slightly higher average spending, with those holding a graduation degree showing the highest average transaction value in the dataset.
>
>However, the difference in spending between education levels is relatively small, suggesting that while education may have some influence, it is not a strong determinant of customer spending behavior in this dataset.

# 5. Conclusion / Key Insights

This exploratory analysis provided insights into the demographic and financial characteristics of credit card customers.

The results show that most customers are over 30 years old, with the largest groups concentrated between 40–49 and 50+ years. In terms of income distribution, the majority of customers earn less than \$40K per year, although a significant portion falls within the \$40K–\$120K range.

The analysis also revealed that customers with Gold cards have the highest average transaction values, while Platinum card users present lower average spending in this dataset.

Additionally, female customers show slightly higher average spending than male customers, although the difference is relatively small. Another interesting finding is that some customers with high credit limits remain inactive for extended periods, which could indicate potential disengagement.

Overall, the dataset demonstrates that customer spending behavior is influenced by multiple factors such as income, credit limit, and card type, rather than a single demographic characteristic.
